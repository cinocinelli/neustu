```python
import numpy as np
import pandas as pd
```

```python
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5)
```

```python
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2)
```

```python
plt.subplot(2,1,1)#指定行和列  (两行一列第一个图)
plt.ylabel('y1 value')
plt.plot(x,y1,'b--')
plt.subplot(2,1,2)#指定行和列  (两行一列第二个图)
plt.ylabel('y2 value')
plt.plot(x,y2,'r--')
plt.show()
```

```python
plt.subplot(221)
plt.plot(x,y1,'b--')
plt.subplot(222)
plt.plot(x,y2,'r--')
plt.subplot(223)
plt.plot(x,y1,'r*')
plt.subplot(224)
plt.plot(x,y2,'y*')
plt.show()
```
