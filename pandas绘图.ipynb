```python
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#series
s1 = Series(np.random.randn(1000))
```

```python
s1
```

```python
s1.cumsum()#前n项累加和
```

```python
s1=s1.cumsum()
```

```python
s1.plot(kind='line',grid=True,label='s1',title='This is s1')
plt.legend()
plt.show()
```

```python
s2= Series(np.random.randn(10))
s1 = Series(np.random.randn(10))
s1.plot(kind='line',grid=True,label='s1')
s2.plot(kind='line',grid=True,label='s2')
plt.show()
```

```python
fig,ax = plt.subplots(2,1)
ax[0].plot(s1)
ax[1].plot(s2)
plt.show()
```

```python
fig,ax = plt.subplots(2,1)
s1.plot(ax=ax[1],label='s1')
s1[:10].plot(ax=ax[0],label='s2',kind='bar')
plt.show()
```
