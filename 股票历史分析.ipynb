```python
import pandas_datareader as pdr
import seaborn as sns
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
start = datetime(2015,9,20)
```

```python
ali = pdr.get_data_yahoo('BABA',start = start)
amzn = pdr.get_data_yahoo('AMZN',start = start)
```

```python
amzn.head()
```

```python
amzn.to_csv('AMZN.csn')
```

```python
ali['Adj Close'].plot(legend = 'True')
plt.show()
```

```python
ali['Volume'].plot(legend = 'True')
plt.show()
```

```python
ali['Adj Close'].plot(legend = 'True')
amzn['Adj Close'].plot(legend = 'True')
plt.show()
```

```python
ali['h-l']=(ali['High']-ali['Low']).plot()
plt.show()
```

```python
#daily return 
ali['dr']=ali['Adj Close'].pct_change()
ali
```

```python
ali['dr'].plot(figsize=(10,4),linestyle='--',marker='o')
plt.show()
```

```python
ali['dr'].plot(kind = 'hist')
plt.show()
```

```python
sns.distplot(ali['dr'].dropna(),bins= 100,color='purple')
plt.show()
```
