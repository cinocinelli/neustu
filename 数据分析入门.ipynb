#  第一个数据分析程序 入门

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x= np.linspace(0,2*np.pi,num=100)#一百个点，等差数列
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
siny=y.copy()
cosy=np.cos(x)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.show()
```
